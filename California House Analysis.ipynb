##What is the average median income of the dataset and check the distribution of data using appropriate plots. Please explain distribution of plot.

```python
#Taking  mean of "Median income" column and define in a variable "avg_med_inc". 
avg_med_inc = round(df['median_income'].mean(),2)
print('Mean of "Median Income" : ',avg_med_inc,'\n')

#Ploting the histogram with x-axis as "Income", y-axis as "Individual Counts"
#And title as "California : Income of People".
plt.xlabel('Income', fontsize = 15)
plt.ylabel('Individuals Count', fontsize = 15)
plt.title('California : Income of People', fontsize = 20)
plt.hist(df['median_income'], color = 'skyblue')


#axvline - Add a vertical line across the axes.
#Drawing the average line and labeling it.
plt.axvline(avg_med_inc, color='white', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(avg_med_inc, max_ylim*0.8, 'Mean: {}'.format(avg_med_inc), color = 'Black')
plt.show()
```

```python
df["median_income"].describe()
```

```python
#IQR - Inter Quartile Range
#IQR = Q3(75%) - Q1(25%)
q3, q1 = 4.7432, 2.5634
IQR = q3 - q1
upper = round((q3 + (1.5 * IQR)),3)
lower = round((q3 - (1.5 * IQR)),3)
print("Outlier :  \nBefore Lower than {}\nAfter Upper than {}".format(lower,upper))
```

#<font color='pink'>**Question - 10**</font>
##Please create a new column named total_bedroom_size. If the total bedrooms are 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise considered to me large.

```python
#Defining an empty list.
total_bedroom_size = []

#Using for loop, taking every value from "total bedrooms".
for i in df['total_bedrooms']:
    #Applying condition for Small, Medium and Large rooms.
    #Appending list from the values.
    if i <= 10:
        total_bedroom_size.append("Small")
    elif (i >=11) and (i < 1000):
        total_bedroom_size.append("Medium")
    else:
        total_bedroom_size.append("Large")

# Defining a new column in "df" and putting all the values of the list.
df["total_bedroom_size"] = total_bedroom_size
df.head()
```
